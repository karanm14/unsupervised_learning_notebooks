```python
import numpy as np
import pandas as pd
from scipy.sparse.linalg import eigs 
import scipy.linalg as lin

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
# Generate clustering data
from sklearn.datasets.samples_generator import make_blobs

k = 2

centers = np.zeros((k,2))
XX, y_true = make_blobs(n_samples=400, centers=k,
                       cluster_std=0.70, random_state=0)

rng = np.random.RandomState(13)

A = np.where(y_true==0)
B = np.where(y_true==1)

XA = np.dot(XX[A,:], rng.randn(2, 2))
rng = np.random.RandomState(13)
XB = np.dot(XX[B,:], rng.randn(2, 2))


X = np.zeros((400,2))

X[0:200,:] = XA[:,:]
X[200:400,:] = XB[:,:]
```

```python
r = np.random.randint(400, size=k)


centers[0,:] = [-2,2.5]
centers[1,:] = [-1,3]



print(centers)
```

```python
n_iter = 20
n = 400
c = np.zeros((n,1))
d = np.zeros((k,1))

for m in range(n_iter):
    A = np.where(c==0)
    B = np.where(c==1)
    scatter = go.Scatter(
    x = X[A[0],0],
    y = X[A[0],1],
    mode='markers',
    marker=dict(
            size=5,
            symbol='circle',
            color='blue',
            opacity=1
        )
    )
    scatter1 = go.Scatter(
        x = X[B[0],0],
        y = X[B[0],1],
        mode='markers',
        marker=dict(
            size=5,
            symbol='circle',
            color='red',
            opacity=1
        )
    )

    
    scatter4 = go.Scatter(
        x = centers[:,0],
        y = centers[:,1],
        mode='markers',
        marker=dict(
            size=8,
            symbol='star',
            color='black',
            opacity=1
        )
    )
    # Above code is for scatter plot of every iteration.

    data = [scatter,scatter1,scatter4]
    iplot(data, filename='fig0')
    # Assign clusters
    for i in range(n):
        d[0] = (X[i,:] - centers[0,:])@(X[i,:] - centers[0,:]).T
        d[1] = (X[i,:] - centers[1,:])@(X[i,:] - centers[1,:]).T
        c[i] = np.argmin(d)
    data = []
    
    # Compute Means and ellipsoids
    A = np.where(c==0)
    centers[0,:] = np.mean(X[A[0],:],0)
    A = np.where(c==1)
    centers[1,:] = np.mean(X[A[0],:],0)
        
        
    

        
    
```
