```python
import numpy as np
import pandas as pd
from scipy.sparse.linalg import eigs 
import scipy.linalg as lin

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
# Generate clustering data
from sklearn.datasets.samples_generator import make_blobs
from scipy.stats import multivariate_normal

k = 2

centers = np.zeros((k,2))

XX, y_true = make_blobs(n_samples=400, centers=k,
                       cluster_std=1, random_state=2)

rng = np.random.RandomState(13)

A = np.where(y_true==0)
B = np.where(y_true==1)

XA = np.dot(XX[A,:], rng.randn(2, 2))
rng = np.random.RandomState(13)
XB = np.dot(XX[B,:], rng.randn(2, 2))


X = np.zeros((400,2))

X[0:200,:] = XA[:,:]
X[200:400,:] = XB[:,:]

```

```python

centers[0,:] = [-2,2.5]
centers[1,:] = [-1,3]

pi = [0.5,0.5]

Sigma1 = np.identity(2)
Sigma2 = np.identity(2)


print(centers)
```

```python
n_iter = 50
n = 400
c = np.zeros((n,1))
d = np.zeros((k,1))
Q = np.zeros((n,2))

for m in range(n_iter):
    c = np.argmax(Q,1)
    A = np.where(c==0)
    B = np.where(c==1)
    scatter = go.Scatter(
    x = X[A[0],0],
    y = X[A[0],1],
    mode='markers',
    marker=dict(
            size=5,
            symbol='circle',
            color='blue',
            opacity=1
        )
    )
    scatter1 = go.Scatter(
        x = X[B[0],0],
        y = X[B[0],1],
        mode='markers',
        marker=dict(
            size=5,
            symbol='circle',
            color='red',
            opacity=1
        )
    )

    
    scatter4 = go.Scatter(
        x = centers[:,0],
        y = centers[:,1],
        mode='markers',
        marker=dict(
            size=8,
            symbol='star',
            color='black',
            opacity=1
        )
    )
    # Above code is for scatter plot of every iteration.

    data = [scatter,scatter1,scatter4]
    iplot(data, filename='fig0')
    # Assign clusters
    print(Sigma1)
    print(Sigma2)
    for i in range(n):
        #d[0] = (X[i,:] - centers[0,:])@Sigma1@(X[i,:] - centers[0,:]).T/2 - np.log(pi[0])
        #d[1] = (X[i,:] - centers[1,:])@Sigma2@(X[i,:] - centers[1,:]).T/2 - np.log(pi[1])
        #c[i] = np.argmin(d)

        var1 = multivariate_normal(mean=centers[0,:], cov=Sigma1)
        var2 = multivariate_normal(mean=centers[1,:], cov=Sigma2)
        Q[i,0] = var1.pdf(X[i,:])*pi[0]
        Q[i,1] = var2.pdf(X[i,:])*pi[1]
        sm =  Q[i,1] + Q[i,0]
        Q[i,0] = Q[i,0]/sm
        Q[i,1] = Q[i,1]/sm
    data = []
    
    # Compute Means and ellipsoids
    centers[0,:] = X.T@Q[:,0]/sum(Q[:,0])
    centers[1,:] = X.T@Q[:,1]/sum(Q[:,1])
    Sigma1 = np.zeros((2,2))
    Sigma2 = np.zeros((2,2))
    for t in range(n):
        Sigma1 = Sigma1 + Q[t,0]*np.asmatrix(X[t,:] - centers[0,:]).T@np.asmatrix(X[t,:] - centers[0,:])/sum(Q[:,0])
        Sigma2 = Sigma2 + Q[t,1]*np.asmatrix(X[t,:] - centers[1,:]).T@np.asmatrix(X[t,:] - centers[1,:])/sum(Q[:,1])
    pi[1] =  sum(Q[:,1])/n
    pi[0] = sum(Q[:,0])/n
        
        
    

        
    
```
