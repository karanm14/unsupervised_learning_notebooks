```python
# Set up the trnasition probability and emission and initial state distribution tables
T = np.zeros((25,25))

pp = 0.95


T[0,1] = 1/2
T[0,5] = 1/2
T[4,3] = 1/2
T[4,9] = 1/2
T[12,7] = 1/4
T[12,11] = 1/4
T[12,13]= 1/4
T[12,17] = 1/4
T[1,0] = 1/3
T[1,2] = 1/3
T[1,6] = 1/3
T[2,1] = 1/3
T[2,3] = 1/3
T[2,7] = 1/3
T[3,2] = 1/3
T[3,4] = 1/3
T[3,8] = 1/3
T[5,0] = 1/3
T[5,6] = 1/3
T[5,10] = 1/3
T[6,1] = (1-pp)/3
T[6,5] = (1-pp)/3
T[6,7] = pp
T[6,11] = (1-pp)/3
T[7,2] = pp/3
T[7,6] = (1-pp)/3
T[7,8] = pp
T[7,12] = (1-pp)/3
T[8,7] = (1-pp)/3
T[8,9] = (1-pp)/3
T[8,3] = (1-pp)/3
T[8,13] = pp
T[9,8] = 1/3
T[9,14] = 1/3
T[9,4] = 1/3
T[10,5] = 1/3
T[10,11] = 1/3
T[10,15] = 1/3
T[11,10] = (1-pp)/3
T[11,12] = (1-pp)/3
T[11,6] = pp
T[11,16] = (1-pp)/3
T[13,8] = (1-pp)/3
T[13,12] = (1-pp)/3
T[13,14]= (1-pp)/3
T[13,18] = pp
T[14,9] = 1/3
T[14,13] = 1/3
T[14,19] = 1/3
T[15,10] = 1/3
T[15,16] = 1/3
T[15,20] = 1/3
T[16,15] = (1-pp)/3
T[16,11] = pp
T[16,21] = (1-pp)/3
T[16,17] = (1-pp)/3
T[17,16] = pp
T[17,18] = (1-pp)/3
T[17,12] = (1-pp)/3
T[17,22] = (1-pp)/3
T[18,17] = pp
T[18,19] = (1-pp)/3
T[18,13] = (1-pp)/3
T[18,23] = (1-pp)/3
T[19,14] = 1/3
T[19,24] = 1/3
T[19,18] = 1/3
T[20,15]=1/2
T[20,21] = 1/2
T[21,20]= 1/3
T[21,22] = 1/3
T[21,16] = 1/3
T[22,21] = 1/3
T[22,23] = 1/3
T[22,17] = 1/3
T[23,18] = 1/3
T[23,22] = 1/3
T[23,24] = 1/3
T[24,19]=1/2
T[24,23] = 1/2

T = 0.95*T + 0.05*np.identity(25);


grid = np.zeros((25,2))

t = 0
for i in range(5):
    for j in range(5):
        grid[t,0] = i
        grid[t,1] = j
        t = t+1

I = np.ones((1,25))/25
```

```python
# This is a function to draw from a given distribiution

def draw(p):
    r = np.random.rand(1,1)
    n = 25
    l = p[0,0];
    
    for i in range(n):
        if(r<=l):
            return i
        else:
            l = l + p[0,i+1]

    return i
```

```python
# We run the bot for n steps (but dont do the plot so we can get things done fast)

n = 2500


p = np.zeros((1,25))
st = draw(I)



x = np.floor(st/5)
y = st - x*5

Dat = np.zeros((n+1,2))
State = np.zeros((n+1,1))
Dat[0,0] = x
Dat[0,1] = y
State[0,0] = st

Obs = np.zeros((n+1,2))
Obs[0,0] = np.sqrt(x*x + y*y) + np.random.randn(1)*0.4
Obs[0,1] = np.sqrt((x-5)*(x-5) + y*y) + np.random.randn(1)*0.4

for i in range(n):
    x = np.floor(st/5)
    y = st - x*5
    p[0,:] = T[st,:]
    st = draw(p)
    Dat[i+1,0] = x
    Dat[i+1,1] = y
    Obs[i+1,0] = np.sqrt(x*x + y*y) + np.random.randn(1)*0.4
    Obs[i+1,1] = np.sqrt((x-5)*(x-5) + y*y) + np.random.randn(1)*0.4
    State[i+1,0] = st


```

```python
# Load and fit HMM to given observation

from hmmlearn import hmm
model = hmm.GaussianHMM(25, "full")
model.fit([Obs])
```

```python
# PRedict most likely sequence of states
Z2 = model.predict(Obs)
```

```python
# Lets print the actual and predicted states side by side.
A = State[:,0]
B = Z2[:]
V= np.zeros((n+1,2))
V[:,0]= A
V[:,1] = B
print(V[200:230,:])
```
